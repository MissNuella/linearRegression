```python
#power transformation
image = cv2.imread(r"C:\Users\EmmanuellaDanso\OneDrive\Pictures\0a777876b73ae120809cf8ef2694a4fe.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image_float = image.astype(np.float32) / 255.0    

gamma_values = np.arange(1, 3, 1)
c_value = np.linspace(1, 4, num=len(gamma_values))

#power transformations; s = crexpy
powered_images = []
for i in gamma_values:
    for c in c_value:
        image_power = c * image_float ** i
        powered_images.append(image_power)


plt.figure(figsize=(15, 5))
num_images = len(powered_images)

for idx, img in enumerate(powered_images):
    plt.subplot(1, num_images, idx + 1)
    plt.imshow(img, vmin=0, vmax=1)
    plt.axis('off')

plt.suptitle("Power Law Transformations")
plt.show()


#gamma correction transformation; s=r* 1/y
corrected_img = []
gamma_vals = np.arange(1, 4, 1)
num = len(gamma_vals)

for i in gamma_vals:
    image_power = powered_images[0] ** (1 / i)
    corrected_img.append(image_power)

    plt.figure(figsize=(15, 3))
    #plt.subplot(1, 3, i + 1)
    plt.imshow(corrected_img[0], cmap='gray')
    plt.axis('off')
```

```python
#gamma correction of the original image like the monitors 
image = cv2.imread(r"C:\Users\EmmanuellaDanso\OneDrive\Pictures\0a777876b73ae120809cf8ef2694a4fe.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Define the gamma value for your monitor (common value is around 2.2
gamma = 2.2

image_float = image.astype(np.float32) / 255.0

# Apply gamma correction: pre-process the image
gamma_corrected = image_float ** (1 / gamma)


# Now, display both images:
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(image, vmin=0, vmax=1)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gamma_corrected, vmin=0, vmax=1)
plt.title('Gamma Corrected Image')
plt.axis('off')

plt.suptitle('Gamma Correction Example')
plt.show()
```

```python
image = cv2.imread(r"C:\Users\EmmanuellaDanso\OneDrive\Pictures\0a777876b73ae120809cf8ef2694a4fe.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image_float = image.astype(np.float32) / 255.0    

gamma_values = np.arange(1, 3, 1)
c_value = np.linspace(1, 4, num=len(gamma_values))

#power transformations; s = crexpy
powered_images = []
for i in gamma_values:
    for c in c_value:
        image_power = c * image_float * np.exp(i)
        powered_images.append(image_power)


plt.figure(figsize=(15, 5))
num_images = len(powered_images)

for idx, img in enumerate(powered_images):
    plt.subplot(1, num_images, idx + 1)
    plt.imshow(img, vmin=0, vmax=1)
    plt.axis('off')

plt.suptitle("Power Law Transformations")
plt.show()
```
